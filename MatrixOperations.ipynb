```python
import numpy as np
```

##### Create two matrices to then solve using numpy function.

###### The algorithm uses LU decomposition to determine the solutions of x. - ??

```python
solve_A = np.array([[1, 3, -1], [2, -1, 1], [-1, 1, -2]])
solve_y = np.array([[-6], [1], [-4]])
```

##### Compute the solution and then print it out.

```python
print(f"We want to solve: \n{solve_A}\n with the vector \n{solve_y}\n ")

x = np.linalg.solve(solve_A, solve_y)
print(f"The solution is: \n{x}\n ")
```

##### This result matches the one I did by hand using Gaussian Elimination. Now, let's go ahead and compute various matrix operations.

```python
A = np.array([[1, 0, 7], [8, 4, 1], [0, 1, 0]])
B = np.array([[-1, 2, 3], [0, 1, -1], [3, 1, 4]])
C = np.array([[1], [2], [3]])
```

```python
print(f"Matrix A is: \n{A}\n ")
print(f"Matrix B is: \n{B}\n ")
```

##### We'll use various numpy functions to determine the product and determinant as well as other operations.

```python
print(f"Product of Matrix AB: \n{np.dot(A, B)}\n ")

print(f"The determinant of A is: \n{np.linalg.det(A)}\n and the determinant of B is: \n{np.linalg.det(B)}\n ")
```

##### Let's see if the inverses exist for Matrix A and B.

```python
print(f"Inverse of A: \n{np.linalg.inv(A)}\n Inverse of B: \n{np.linalg.inv(B)}")
```

##### Great! We can see that the inverses do exist. Let's determine the rank of each matrix.

###### The rank of a matrix will tell us about whether a vector is linearly independent or dependent.

```python
print(f"Rank of A is: \n{np.linalg.matrix_rank(A)}\n Rank of B is: \n{np.linalg.matrix_rank(B)}\n ")
```

##### Now, we need to figure out if the vector y is a linear combination of the columns of matrix A.

###### We can do this by checking the rank of an augmented matrix of A and y (which is C). 

```python
if np.linalg.matrix_rank(np.concatenate((A, C), axis=1)) == np.linalg.matrix_rank(A) + 1:
    print(f"The system of equations is consistent and has a unique solution.")
else:
    print(f"The system of equations is inconsistent and has no solution.")
```

##### Thus, we have our answer that there is no solution with the addition of our vector y.
