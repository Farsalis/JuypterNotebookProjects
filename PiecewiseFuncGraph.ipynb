```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Create a function that acts mathematically similar to the piecewise function.

This part is not too difficult, branching if statements act like the conditional intervals in a piecewise function.

```python
def generate_y_values(x_arr, y_arr):
    """Generate a numpy array of points based on function of x."""
    for i in range(len(x_arr)): # Apply x input to get y output
        if x_arr[i] < -1:
            y_arr[i] = x_arr[i]**2
        elif x_arr[i] <= 2:
            y_arr[i] = x_arr[i]**3
        else:
            y_arr[i] = x_arr[i]**4
    return y_arr
```

#### Initialize numpy array for values of x and call function to generate values of y.

We use x_values as a base for y_values by zeroing each item in it. Range of values of x matter, otherwise the non-differentiable points cannot be seen.

```python
x_values = np.linspace(-4, 3,1000)
y_values = generate_y_values(x_values, np.zeros_like(x_values))  # Function which changes all values of x to zero.
```

#### Now, create the graph via matplotlib and check the result.

tight_layout() center and fixes the spacing of our graph. .scatter() just let's us plot points. Additionally, we can control our graph size via .figure().

```python
plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, 'g--', linewidth=1, alpha=0.9)
plt.scatter([-1, 2], [1, 8], color='darkred', s=80, zorder=5)
plt.grid(True, alpha=0.3, linewidth=0.7)
plt.title(f"Graph of Piecewise Function")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.tight_layout()
plt.show()
```

#### It turned out pretty well. It's clear that it is a piecewise function.
