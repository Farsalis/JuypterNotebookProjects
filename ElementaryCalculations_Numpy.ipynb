```python
import numpy as np
from collections.abc import Callable
```

#### Use numpy as it has more accurate pi value and test function.

##### You can use numpy to do the calculation like those used below but I wanted to show both ways.

```python
def calc_circumference(r:float):
    pi=np.pi
    return 2*pi*r, pi*r**2
```

```python
help(calc_circumference)
```

```python
area, circumference = calc_circumference(5)
print(f"Computation Done: Area: {area}, Circumference: {circumference}")

area, circumference = calc_circumference(10)
print(f"Second Computation Done: Area: {area}, Circumference: {circumference}")
```

#### Define the function which computes arbitrary f and g.

```python
def calc_funcs(f:Callable, g:Callable, x:float):
    first_computation = f(g(x))
    second_computation = g(f(x))
    return first_computation, second_computation
```

#### You could also define the functions directly in the inputs. Use numpy for more efficient calculations.

##### This result is incredibly interesting. Mathematically these results are the same but are different here. 

```python
square = lambda x: np.square(x)
square_root = lambda x: np.sqrt(x)

first_comp, second_comp = calc_funcs(square, square_root, x=3.0)
print(f"First computation: {first_comp}, Second computation: {second_comp}")
```

#### Let's try some other numbers. 

##### So the result is dependent on our number. It seems like whether it is prime doesn't matter. Divisibility by 2 doesn't seem to matter too.

```python
for i in range(1, 15):
    first_comp, second_comp = calc_funcs(square, square_root, i)
    print(f"First computation: {first_comp}, Second computation: {second_comp}")
```
