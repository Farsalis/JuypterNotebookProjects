```python
import numpy as np

list_example = [1, 2, 3, 4, 5]

string_example = "This is a string wow!"

array_example = np.array(list_example)
```

### No need to print twice, use function instead. 

```python
def call_print(message: str=""):
    print(f"List {message}: {list_example[1]}, {list_example}")
    print(f"Array {message}: {array_example.item(len(list_example) - 2)}, {array_example}")
```

```python
print(string_example[2], string_example.replace("i", "I",2))

call_print("Original")

array_example[array_example.size-2]=3
list_example[1]=1

call_print("Modified")
```
