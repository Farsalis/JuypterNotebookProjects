```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import polynomial as poly
```

##### Define the factorial function recursively and create a file to output the result.

We also raise an error if our number is not a positive integer. If the functions fails for any other reason we raise a general recursion error (this could happen for any reason).

```python
ans_file = open('ans.txt', 'w')

def compute_factorial(n: int):
    """
    Recursively computer the factorial of a number.

    :param n:
    :return:
    """
    try:
        if n < 0:
            raise TypeError("Factorial is not defined for negative numbers")
        elif n == 0:
            return 1
        else:
            return n * compute_factorial(n - 1)
    except RecursionError:
        print("Recursive function has failed. Please try again.")
        return None
```

Test the function to make sure it is correct.

```python
print(f"Result: {compute_factorial(10)}")

print(f"Result: {compute_factorial(5)}")
```

##### The output looks correct so let's write it to the file.

```python
ans_file.write(f"10! = {compute_factorial(10)}\n")
ans_file.write(f"5! = {compute_factorial(5)}\n")
ans_file.write("Samuel Frerichs.\n")
ans_file.close()
```

##### Define a function to compute polynomials via numpy's function.

.linspace() generates a hundred points in the given domain and .Polynomial() instances the polynomial as a python object (so we have extra methods and features that can be used).

##### .Polynomial() acts like a regular mathematical function, so we can input x_values into it and get y_values.

*args are any extra parameters user wants to put through the function.

```python
def compute_polynomial(arr, *args):
    try:
        x_values = np.linspace(-5, 5, 100)
        y_values = poly.Polynomial(arr, *args)(x_values)
        return x_values, y_values
    except TypeError as e:
        print(f"Error: {e}\n Make sure you are using an array of integers.")
        return None, None
```

##### Create the plot and customize via matplotlib functions.

```python
def graph_polynomial():
    x, y = compute_polynomial([4, 5, 6])
    x_2, y_2 = compute_polynomial([1,8,2,])
    x_3, y_3, = compute_polynomial([1,2,3])
    plt.plot(x, y, 'r-', label="Trinomial")
    plt.plot(x_2, y_2, 'g-', label="5-Term Polynomial")
    plt.plot(x_3, y_3, 'b-', label="8-Term Polynomial")

    plt.xlabel("x-axis")
    plt.ylabel("y-axis")

    plt.legend(loc="upper left")
    plt.title("Polynomial Graph")

    plt.show()

graph_polynomial()
```
